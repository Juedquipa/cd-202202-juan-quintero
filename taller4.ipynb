```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib import colors
from scipy.stats import pearsonr, zscore

import statsmodels.api as sm
import statsmodels.formula.api as smf

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error, r2_score
from sklearn import linear_model
```

```python
#Datos de hoteles
ruta = 'https://raw.githubusercontent.com/rpizarrog/FundamentosMachineLearning/master/datos/melb_data.csv'
df = pd.read_csv(ruta)
```

# Building area vs YearBuilt (Rooms)

```python
df0 = df.copy(deep=True)
```

```python
df0 = df0[df0['BuildingArea'].notna()]
df0 = df0[df0['YearBuilt'].notna()]
```

```python
df0 = df0[(np.abs(zscore(df0['BuildingArea']))<=3)]
df0 = df0[(np.abs(zscore(df0['YearBuilt']))<=3)]
```

```python
#Diagrama de distribución
fig, ax = plt.subplots(figsize=(10, 6))
classes = df0['Rooms'].to_list()
scatter = plt.scatter(x=df0['BuildingArea'], y=df0['YearBuilt'], c=classes, cmap='Wistia')
ax.set_title('Distribución de área construida y año de construcción')
ax.legend(*scatter.legend_elements())
```

Se puede ver que el área construida influye bastante en el número de habitaciones que hay en la vivienda, además se puede notar que entre más antigua es la vivienda, menos probable es que tenga un gran parea construida.

# Car vs Bedroom2 (Rooms)

```python
df1 = df.copy(deep=True)
```

```python
df1 = df1[df1['Car'].notna()]
df1 = df1[df1['Bedroom2'].notna()]
```

```python
#Diagrama de distribución
fig, ax = plt.subplots(figsize=(10, 6))
classes = df1['Rooms'].to_list()
scatter = plt.scatter(x=df1['Car'], y=df1['Bedroom2'], c=classes, cmap='copper_r')
ax.set_title('Distribución de garajes/carros y bedrooms')
ax.legend(*scatter.legend_elements())
```

Se puede observar que entre más garajes, más habitaciones suele tener el edificio, aunque la escala de garaje vs habitaciones pareciera seguir un tendencia un poco logaritmica.
